ZADANIE 1: Analiza danych sprzedażowych
---------------------------------------
Dane: Miesięczna sprzedaż produktu w PLN

sprzedaz = [12500, 13200, 15800, 14200, 16500, 18200, 17800, 19200,
            20500, 21300, 23500, 25200]

a) Oblicz podstawowe statystyki (średnia, mediana, min, max)

b) Znajdź miesiące ze sprzedażą powyżej średniej

c) Oblicz procentowy wzrost sprzedaży (grudzień vs styczeń)

d) Stwórz wykres liniowy pokazujący trend

e) Dodaj linię średniej na wykresie


````python
import numpy as np
import matplotlib.pyplot as plt

# a) Statystyki

sprzedaz = [12500, 13200, 15800, 14200, 16500, 18200, 17800, 19200, 20500, 21300, 23500, 25200]
srednia = np.mean(sprzedaz)
mediana = np.median(sprzedaz)
minimum = np.min(sprzedaz)
maksimum = np.max(sprzedaz)

print(f"Średnia: {srednia:.2f} PLN")
print(f"Mediana: {mediana:.2f} PLN")
print(f"Min: {minimum} PLN, Max: {maksimum} PLN")

# b) Miesiące powyżej średniej
miesiace = ['Sty', 'Lut', 'Mar', 'Kwi', 'Maj', 'Cze',
            'Lip', 'Sie', 'Wrz', 'Paź', 'Lis', 'Gru']

powyzej_sredniej = [m for m, v in zip(miesiace, sprzedaz) if v > srednia]
print("Miesiące ze sprzedażą powyżej średniej:", ", ".join(powyzej_sredniej))

# c) Procentowy wzrost (grudzień vs styczeń)
wzrost = (sprzedaz[-1] - sprzedaz[0]) / sprzedaz[0] * 100
print(f"Procentowy wzrost sprzedaży (Gru vs Sty): {wzrost:.2f}%")

# d) Wykres trendu
plt.figure()
plt.plot(miesiace, sprzedaz, marker='o', label='Sprzedaż (PLN)')
plt.axhline(srednia, linestyle='--', label=f"Średnia = {srednia:.0f}")
plt.title("Miesięczna sprzedaż produktu")
plt.xlabel("Miesiąc")
plt.ylabel("Sprzedaż [PLN]")
plt.legend()
plt.grid(True)
plt.show()
````

ADANIE 2: Słowniki - Analiza portfolio
---------------------------------------
Stwórz słownik opisujący portfolio inwestycyjne:
```
portfolio = {
    'AAPL': {'liczba': 10, 'cena_zakupu': 150, 'cena_aktualna': 175},
    'GOOGL': {'liczba': 5, 'cena_zakupu': 2800, 'cena_aktualna': 2950},
    'MSFT': {'liczba': 15, 'cena_zakupu': 300, 'cena_aktualna': 320}
}
```

a) Oblicz całkowitą wartość początkową portfolio

b) Oblicz aktualną wartość portfolio

c) Oblicz zysk/stratę dla każdej akcji

d) Które akcje przynoszą największy zwrot (w %)?


````python
portfolio = {
    'AAPL': {'liczba': 10, 'cena_zakupu': 150, 'cena_aktualna': 175},
    'GOOGL': {'liczba': 5, 'cena_zakupu': 2800, 'cena_aktualna': 2950},
    'MSFT': {'liczba': 15, 'cena_zakupu': 300, 'cena_aktualna': 320}
}

# a) Całkowita wartość początkowa
wartosc_poczatkowa = sum(v['liczba'] * v['cena_zakupu'] for v in portfolio.values())

# b) Aktualna wartość
wartosc_aktualna = sum(v['liczba'] * v['cena_aktualna'] for v in portfolio.values())

# c) Zysk/strata dla każdej akcji
zyski = {ticker: (info['cena_aktualna'] - info['cena_zakupu']) * info['liczba']
         for ticker, info in portfolio.items()}

print(f"Wartość początkowa portfela: {wartosc_poczatkowa:.2f} USD")
print(f"Aktualna wartość portfela:   {wartosc_aktualna:.2f} USD")
print("Zysk/strata dla każdej pozycji:")
for t, z in zyski.items():
    print(f"  {t}: {z:+.2f} USD")

caly_zysk = wartosc_aktualna - wartosc_poczatkowa
print(f"Łączny zysk portfela: {caly_zysk:+.2f} USD")

# d) Zwrot procentowy i najlepsza akcja
zwroty_proc = {ticker: ((info['cena_aktualna'] - info['cena_zakupu']) / info['cena_zakupu']) * 100
               for ticker, info in portfolio.items()}
najlepsza = max(zwroty_proc, key=zwroty_proc.get)

print("Zwroty procentowe:")
for t, p in zwroty_proc.items():
    print(f"  {t}: {p:.2f}%")
print(f"Najlepsza inwestycja: {najlepsza} ({zwroty_proc[najlepsza]:.2f}%)")
````

ZADANIE 3: Wizualizacja danych klimatycznych
--------------------------------------------
temp_2022 = [2, 4, 8, 13, 18, 21, 24, 23, 19, 14, 8, 3]


temp_2023 = [3, 5, 9, 14, 19, 22, 25, 24, 20, 15, 9, 4]

miesiace = ['Sty', 'Lut', 'Mar', 'Kwi', 'Maj', 'Cze',
            'Lip', 'Sie', 'Wrz', 'Paź', 'Lis', 'Gru']

a) Stwórz wykres liniowy porównujący temperatury w obu latach

b) Zaznacz najcieplejszy i najzimniejszy miesiąc każdego roku

c) Oblicz średnią różnicę temperatur między latami

d) Dodaj odpowiednie etykiety, legendę i tytuł

````python
import numpy as np
import matplotlib.pyplot as plt

temp_2022 = [2, 4, 8, 13, 18, 21, 24, 23, 19, 14, 8, 3]
temp_2023 = [3, 5, 9, 14, 19, 22, 25, 24, 20, 15, 9, 4]
miesiace = ['Sty', 'Lut', 'Mar', 'Kwi', 'Maj', 'Cze',
            'Lip', 'Sie', 'Wrz', 'Paź', 'Lis', 'Gru']

# a) Wykres liniowy porównujący
plt.figure()
plt.plot(miesiace, temp_2022, marker='o', label='2022')
plt.plot(miesiace, temp_2023, marker='o', label='2023')

# b) Najcieplejszy i najzimniejszy miesiąc każdego roku
max22_idx = int(np.argmax(temp_2022)); min22_idx = int(np.argmin(temp_2022))
max23_idx = int(np.argmax(temp_2023)); min23_idx = int(np.argmin(temp_2023))

plt.scatter([miesiace[max22_idx]], [temp_2022[max22_idx]], s=100, label='Max 2022')
plt.scatter([miesiace[min22_idx]], [temp_2022[min22_idx]], s=100, label='Min 2022')
plt.scatter([miesiace[max23_idx]], [temp_2023[max23_idx]], s=100, label='Max 2023')
plt.scatter([miesiace[min23_idx]], [temp_2023[min23_idx]], s=100, label='Min 2023')

# c) Średnia różnica temperatur (2023 - 2022)
roznice = np.array(temp_2023) - np.array(temp_2022)
srednia_roznica = np.mean(roznice)
print(f"Średnia różnica temperatur (2023 - 2022): {srednia_roznica:.2f}°C")

# d) Opisy
plt.title("Średnie temperatury miesięczne: 2022 vs 2023")
plt.xlabel("Miesiąc")
plt.ylabel("Temperatura [°C]")
plt.legend()
plt.grid(True)
plt.show()


# e) Osobny wykres liniowy różnic temperatur (2023 - 2022)
plt.figure(figsize=(8,5))
plt.plot(miesiace, roznice, marker='o', linestyle='-', color='blue', label='Różnica (2023 - 2022)')
plt.axhline(0, color='red', linestyle='--', label='Brak różnicy')
plt.title("Różnica temperatur między 2023 a 2022 (°C)")
plt.xlabel("Miesiąc")
plt.ylabel("Różnica [°C]")
plt.legend()
plt.grid(True)
plt.show()
````

ZADANIE 4: Symulacja Monte Carlo
---------------------------------
Zasymuluj 1000 rzutów kostką (6-ścienną):

a) Wygeneruj dane używając np.random.randint(1, 7, 1000)

b) Stwórz histogram pokazujący częstość każdej liczby

c) Oblicz teoretyczne i empiryczne prawdopodobieństwo każdej liczby

d) Czy wyniki są zgodne z teorią? (każda liczba ~16.67%)

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)  # dla powtarzalności

# Symulacje
rzuty_1000 = np.random.randint(1, 7, 1000)
rzuty_10000 = np.random.randint(1, 7, 10000)

# Histogram dla 1000 rzutów
plt.figure(figsize=(8,5))
plt.hist(rzuty_1000, bins=np.arange(0.5, 6.6, 1), edgecolor='black')
plt.title("Rozkład wyników - 1000 rzutów kostką")
plt.xlabel("Wynik")
plt.ylabel("Częstość")
plt.grid(True)
plt.show()

# Histogram dla 10000 rzutów
plt.figure(figsize=(8,5))
plt.hist(rzuty_10000, bins=np.arange(0.5, 6.6, 1), edgecolor='black')
plt.title("Rozkład wyników - 10000 rzutów kostką")
plt.xlabel("Wynik")
plt.ylabel("Częstość")
plt.grid(True)
plt.show()

# Funkcja do analizy
def analiza(rzuty, nazwa):
    unikalne, czestosci = np.unique(rzuty, return_counts=True)
    empiryczne = {int(k): v/len(rzuty) for k, v in zip(unikalne, czestosci)}
    teoretyczne = {k: 1/6 for k in range(1,7)}
    odchylenia = {k: abs(empiryczne.get(k, 0) - teoretyczne[k]) for k in range(1,7)}
    srednie_odchylenie = sum(odchylenia.values())/6

    print(f"\n=== {nazwa} ===")
    print("Empiryczne prawdopodobieństwa:")
    for k in range(1,7):
        print(f"  {k}: {empiryczne.get(k, 0):.4f}")
    print(f"Średnie odchylenie od teorii: {srednie_odchylenie:.4f}")
    return srednie_odchylenie

# Wyniki
odch_1000 = analiza(rzuty_1000, "1000 rzutów")
odch_10000 = analiza(rzuty_10000, "10000 rzutów")

print("\nPorównanie:")
print(f"Średnie odchylenie 1000 rzutów:  {odch_1000:.4f}")
print(f"Średnie odchylenie 10000 rzutów: {odch_10000:.4f}")
print("Wniosek: przy większej liczbie prób rozkład lepiej przybliża rozkład teoretyczny (1/6).")

````
