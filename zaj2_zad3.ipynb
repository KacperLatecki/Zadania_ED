```python
# ===============================================
# 1. Import bibliotek
# ===============================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")


# ===============================================
# 2. Wczytanie danych
# (wgraj train.csv do Colab  Files  Upload)
# ===============================================
df = pd.read_csv("train.csv")
df.head()


# ===============================================
# 3. Podstawowy opis danych
# ===============================================
print(df.info())
df.describe()


# ===============================================
# 4. Histogram i KDE  SalePrice
# ===============================================
plt.figure(figsize=(10,5))
sns.histplot(df['SalePrice'], kde=True)
plt.title("Rozkad ceny domu  SalePrice")
plt.show()


# ===============================================
# 5. Boxplot  SalePrice
# ===============================================
plt.figure(figsize=(10,5))
sns.boxplot(x=df['SalePrice'])
plt.title("Boxplot SalePrice")
plt.show()


# ===============================================
# 6. Macierz korelacji  tylko dane numeryczne
# ===============================================

# wybieramy tylko kolumny numeryczne
df_num = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(16,10))
sns.heatmap(df_num.corr(), cmap="coolwarm")
plt.title("Macierz korelacji (cechy numeryczne)")
plt.show()



# ===============================================
# 7. Scatterploty jak na wykadzie
# ===============================================

# GrLivArea vs SalePrice
plt.figure(figsize=(8,6))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df)
plt.title("GrLivArea vs SalePrice")
plt.show()

# OverallQual vs SalePrice
plt.figure(figsize=(8,6))
sns.scatterplot(x='OverallQual', y='SalePrice', data=df)
plt.title("OverallQual vs SalePrice")
plt.show()

# YearBuilt vs SalePrice
plt.figure(figsize=(8,6))
sns.scatterplot(x='YearBuilt', y='SalePrice', data=df)
plt.title("YearBuilt vs SalePrice")
plt.show()


# ===============================================
# 8. Pairplot (jak na wykadzie)
# ===============================================
cols = ['SalePrice', 'GrLivArea', 'OverallQual', 'GarageCars', 'TotalBsmtSF']
sns.pairplot(df[cols], corner=True)
plt.show()


# ===============================================
# 9. Dodatkowe wykresy (opcjonalnie)
# ===============================================

# Liczba pokoi vs Cena
plt.figure(figsize=(8,6))
sns.boxplot(x='TotRmsAbvGrd', y='SalePrice', data=df)
plt.title("TotRmsAbvGrd vs SalePrice")
plt.show()

# HouseStyle vs Cena
plt.figure(figsize=(12,6))
order = df.groupby("HouseStyle")['SalePrice'].median().sort_values().index
sns.boxplot(x="HouseStyle", y="SalePrice", data=df, order=order)
plt.xticks(rotation=45)
plt.title("HouseStyle vs SalePrice")
plt.show()


# ===============================================
# 10. Usunicie outlierw (opcjonalnie)
# ===============================================
df_no_outliers = df[df['GrLivArea'] < 4500]

plt.figure(figsize=(8,6))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df_no_outliers)
plt.title("GrLivArea vs SalePrice  bez outlierw")
plt.show()

# ===============================================
# KONIEC NOTEBOOKA
# ===============================================
```
